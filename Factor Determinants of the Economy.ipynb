## PHASE 1: DATA COLLECTION AND WRANGLING

```python
# Import Libraries for wrangling data
import pandas as pd
import numpy as np
```

The Function below automatically cleans our data set when called, it takes two prameters the filename(.extension) of the dataset and the output column name of your choice.

```python
def clean_data(filename,column_name):
    """This Function automates the data wrangling
    process for our datasets"""
    data = pd.read_csv(filename)
    
    nigeria_data = data.loc[data['Country Name']=='Nigeria',:]
    
    nigeria_data = nigeria_data.drop(['Country Name','Country Code','Indicator Name','Indicator Code','Unnamed: 64'],
                        axis=1,
                       )
    
    year_list = [str(y) for y in range(1960,2020)] 
    nigeria_data = nigeria_data.melt(value_vars=year_list,
                                                 var_name='Year',
                                                 value_name=column_name
                                            )
    return nigeria_data
```

### GDP DATA

Gross Domestic Product (GDP): This study takes the GDP as an important indicator of
economic growth because the GDP concentrates on the output produced within the country.

Data source: https://data.worldbank.org/indicator/NY.GDP.MKTP.CD?locations=NG&view=chart

```python
nigeria_gdp = clean_data('WORLD_GDP.csv','GDP')
nigeria_gdp.tail(10)
```

### POPULATION DATA

## PHASE 3: Multiple Linear Regression Model

<p>Multiple Linear Regression is a method to help us understand the relationship between three or more variables:</p>
<ul>
    <li>The predictors/independent variables (X1,X2,X3,...,Xn)</li>
    <li>The response/dependent variable (that we want to predict)(Y)</li>
</ul>

<p>The result of Linear Regression is a <b>linear function</b> that predicts the response (dependent) variable as a function of the predictors (independent) variables.</p>


$$
Y: Response \ Variable\\
X_1 :Predictor\ Variable \ 1\\
X_2: Predictor\ Variable \ 2\\
X_3: Predictor\ Variable \ 3\\
X_4: Predictor\ Variable \ 4\\
$$

$$
a: intercept\\
b_1 :coefficients \ of\ Variable \ 1\\
b_2: coefficients \ of\ Variable \ 2\\
b_3: coefficients \ of\ Variable \ 3\\
b_4: coefficients \ of\ Variable \ 4\\
$$

The equation is given by

$$
Yhat = a + b_1 X_1 + b_2 X_2 + b_3 X_3 + b_4 X_4
$$

Lets load the modules for linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

Create the linear regression object

```python
lm=LinearRegression()
model=lm.fit(X_train,y_train)
Y_hat=model.predict(X_test)
```

<p>From the previous section  we know that good predictors of GDP are:</p>
<ul>
    <li>POPULATION</li>
    <li>IMPORT RATE</li>
    <li>OIL REVENUE</li>    
</ul>
We would develop a model using these variables as the predictor variables.

Fit the linear model using the 3 above-mentioned variables.

```python
print("Our intercerpt(a) is -> ", lm.intercept_)
```

```python
print("Our Coefficients are -> ",lm.coef_)
```

As we saw above, we should get a final linear function with the structure:

$$
Yhat = a + b_1 X_1 + b_2 X_2 + b_3 X_3 + b_4 X_4
$$

We get

<b>GDP</b> = 51686155240.7632 + 4.87703590e+01 x <b>POPULATION</b> + 6.39844632e+00 x <b>IMPORT RATE</b> + -2.88536504e+09 x <b>OIL REVENUE</b>

Lastly let's make our prediction and use a <b>Distribution Plot</b> to examine the fit of our model

We can look at the distribution of the fitted values that result from the model and compare it to the distribution of the actual values.

```python
print("Train Score:", model.score(X_train,y_train))
```

```python
print("Test Score:", model.score(X_test,y_test))
```

```python
print("R-Squared Score: ",r2_score(y_test,Y_hat))
```

```python
plt.figure(figsize=(12,5));
ax1=sns.distplot(y_test,hist=False,label="Actual GDP Values")
sns.distplot(Y_hat,ax=ax1,hist=False,label="Predicted GDP Values");
```

In conclusion, We can see that the fitted values are reasonably close to the actual values, since the two distributions overlap a bit. However, there is definitely some room for improvement.
